```python
import requests
from bs4 import BeautifulSoup
import zipfile
import os

headers = {
    "User-Agent": "Mozilla/5.0"
}

# ✅ HÀM: Tải âm thanh + Lấy IPA
def get_audio_and_ipa(word):
    url = f"https://dictionary.cambridge.org/dictionary/english/{word.replace(' ', '%20')}"
    response = requests.get(url, headers=headers)

    if response.status_code != 200:
        print(f"❌ Không thể truy cập từ: {word}")
        return "Không có IPA"

    soup = BeautifulSoup(response.text, "html.parser")

    # 📥 Tải âm thanh (giọng Mỹ)
    try:
        audio_tag = soup.find("span", class_="us dpron-i")
        source_tag = audio_tag.find("source") if audio_tag else None
        if source_tag and source_tag.get("src"):
            audio_url = "https://dictionary.cambridge.org" + source_tag["src"]
            audio_response = requests.get(audio_url, headers=headers)
            if audio_response.status_code == 200:
                filename = word.replace(" ", "_") + ".mp3"
                with open(filename, "wb") as f:
                    f.write(audio_response.content)
                print(f"🔊 Đã tải: {filename}")
            else:
                print(f"⚠️ Không tải được âm thanh cho: {word}")
        else:
            print(f"🔇 Không có âm thanh cho: {word}")
    except Exception as e:
        print(f"❗ Lỗi khi tải âm thanh: {word} – {e}")

    # 📘 Lấy phiên âm IPA (giọng Mỹ)
    try:
        ipa_span = soup.find("span", class_="us dpron-i")
        ipa_inner = ipa_span.find("span", class_="pron dpron") if ipa_span else None
        ipa_text = ipa_inner.text.strip() if ipa_inner else "Không có IPA"
    except Exception as e:
        ipa_text = "Không có IPA"

    print(f"📘 {word}: {ipa_text}")
    return ipa_text

# ✅ DANH SÁCH TỪ CỦA BẠN (bao nhiêu từ cũng được)
danh_sach_tu = """
refeshment, combine, array, oatmeal, pastrie, aware, accommodate, forward, congest, solicit, reduction, none, ripe, sauce, hand, guess, pepper, ingredient


"""

# ✅ Xử lý danh sách
ds_tu_vung = [tu.strip().lower() for tu in danh_sach_tu.replace("\n", ",").split(",") if tu.strip()]
ipa_dict = {}

for tu in ds_tu_vung:
    ipa = get_audio_and_ipa(tu)
    ipa_dict[tu] = ipa

# ✅ Lưu phiên âm vào file txt
with open("ipa_list.txt", "w", encoding="utf-8") as f:
    for tu, ipa in ipa_dict.items():
        f.write(f"{tu}: {ipa}\n")
print("📄 Đã lưu phiên âm vào ipa_list.txt")

# ✅ Nén tất cả file .mp3 vào file .zip
def zip_mp3_files(zip_name="audio_files.zip"):
    with zipfile.ZipFile(zip_name, 'w') as zipf:
        for file in os.listdir():
            if file.endswith(".mp3"):
                zipf.write(file)
    print(f"🗜️ Đã nén vào: {zip_name}")

zip_mp3_files()
```
